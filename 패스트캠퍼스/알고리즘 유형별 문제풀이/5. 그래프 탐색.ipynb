```python
from collections import deque, defaultdict
n, m, start = map(int, input().split())

graph = defaultdict(list)
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
# BFS
visited = [0 for _ in range(n+1)]
queue = deque()
queue.append(start)
bfs = []
while queue:
    temp = queue.popleft()
    if visited[temp] == 0:
        bfs.append(temp)
        visited[temp] = 1
        graph[temp].sort()
        queue.extend(graph[temp])
        
# DFS
# stackì„ ì´ìš©í•´ë„ ë˜ì§€ë§Œ ì¬ê·€í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ê² ìŒ
# ìê¸°ë‘ ì—°ê²°ëœì•  dfs í˜¸ì¶œ, í˜¸ì¶œ 
visited = [0 for _ in range(n+1)]
def dfs(start):
    print(start, end = ' ')
    visited[start] = 1
    for item in graph[start]:
        if visited[item] == 0:
            dfs(item)
            

dfs(start)
print()
for item in bfs:
    print(item, end = ' ')
    
```

- ì—¬ê¸°ì„  bfsìˆœì„œê°€ ì¤‘ìš”í•œê²Œ ì•„ë‹ˆë‹ˆê¹Œ ê·¸ëƒ¥ stack, dfsë¡œ êµ¬í˜„í•˜ì
- dequeë¥¼ ì‚¬ìš©í•˜ëŠ”ê²Œ pop(0)ë³´ë‹¨ ì‹œê°„ì´ ëœê±¸ë¦¬ì§€ë§Œ pop()ë§Œí¼ì€ ì•„ë‹ˆë‹¤!

```python
# from collections import deque
testCase = int(input())
for _ in range(testCase):
    m, n, k = map(int, input().split())
    table = [[0] * n for _ in range(m)]
    array = []
    for _ in range(k):
        a, b = map(int, input().split())
        array.append((a,b))
        table[a][b] = 1

    def bfs(x, y):
        queue = []
        queue.append((x,y))
        while queue:
            a, b = queue.pop()
            table[a][b] = 0
            if a-1 >= 0 and table[a-1][b] == 1:
                queue.append((a-1, b))
            if a+1 < m and table[a+1][b] == 1:
                queue.append((a+1, b))
            if b-1 >= 0 and table[a][b-1] == 1:
                queue.append((a, b-1))
            if b+1 < n and table[a][b+1] == 1:
                queue.append((a, b+1))
    cnt = 0
    for x, y in array:
        if table[x][y] == 1:
            bfs(x, y)
            cnt += 1
    print(cnt)
```

```python
import heapq
test_case = int(input())
for _ in range(test_case):
    n, d, c = map(int, input().split())
    graph = [[] for _ in range(n+1)]
    for i in range(d):
        a, b, s = map(int, input().split())
        graph[b].append((s, a))

    queue = []
    heapq.heappush(queue, (0, c))
    distances = [float('inf') for _ in range(n+1)]
    distances[c] = 0
    while queue:
        cS, cNode = heapq.heappop(queue)
        if cS <= distances[cNode]:
            for s, a in graph[cNode]:
                dis = cS + s
                if distances[a] > dis:
                    distances[a] = dis
                    heapq.heappush(queue, (dis, a))

    cnt = 0
    time = 0
    for item in distances:
        if item != float('inf'):
            cnt += 1
            if time < item:
                time = item
    print(cnt, time)
```

```python
import math
def get_distance(p1, p2):
    result = (p1[0]-p2[0])**2 + (p1[1]-p2[1])**2
    return math.sqrt(result)

def find(node):
    if node != parent[node]:
        parent[node] = find(parent[node])
    return parent[node]
    
def union(u, v):
    root1 = find(u)
    root2 = find(v)
    if rank[root1] > rank[root2]:
        parent[root2] = root1
    else:
        parent[root1] = root2
        if rank[root1] == rank[root2]:
            rank[root2] += 1
    

n,m = map(int, input().split())
vertex = [i for i in range(1, n+1)]
location = [[] for _ in range(n+1)]
for i in range(1, n+1):
    x, y = map(int,input().split())
    location[i] =[x,y]

# ëª¨ë“  ë…¸ë“œì— ëŒ€í•˜ì—¬ ê°€ëŠ¥í•œ ê°„ì„  ì¶”ê°€ 
edges = []
for i in range(1, n):
    for j in range(i+1, n+1):
        edges.append((get_distance(location[i], location[j]), i, j))

# ì´ˆê¸°í™” 
parent = dict()
rank = dict()
for item in vertex:
    parent[item] = item
    rank[item] = 0

#ì •ë ¬
edges.sort()

for i in range(m):
    a, b = map(int, input().split())
    union(a, b)
    
# ì—°ê²°
result = 0
for edge in edges:
    distance, u, v = edge
    if find(u) != find(v):
        result += distance
        union(u, v)
print("%0.2f"%result)
```
