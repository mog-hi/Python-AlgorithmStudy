```python
def isAscending(data):
    check = 0;
    for i in range(1,9):
        if int(data[i-1]) == i:
            check += 0
        else: 
            check += 1
    if check == 0:
        return True
    else:
        return False
    
def isDescending(data):
    check = 0;
    for i in range(1,9):
        if int(data[8-i]) == i:
            check += 0
        else: 
            check += 1
    if check == 0:
        return True
    else:
        return False
    

a = input().split(' ')
if isAscending(a):
    print("ascending")
elif isDescending(a):
    print("descending")
else:
    print("mixed")
```

```python
a = list(map(int, input().split()))

ascending = True
descending = True
# í•˜ë‚˜ë¼ë„ í‹€ë¦¬ë©´ ì•„ë‹Œ ìƒí™©ì—ì„œëŠ” ì´ˆê¸°ê°’ì„ trueë¡œ ë‘ê³  í‹€ë¦¬ë©´ falseì²´í¬í•˜ëŠ” ê²ƒì´ í¸í•˜ë‹¤.
# ì˜¤ë¦„ì°¨ìˆœ, ë‚´ë¦¼ì°¨ìˆœ ê²€ì‚¬
for i in range(1, 8):
    if a[i-1] < a[i]:
        descending = False
    if a[i-1] > a[i]:
        ascending = False

if ascending:
    print("ascending")
elif descending:
    print("descending")
else:
    print("mixed")
```

```python
n,m = list(map(int, input().split()))

card = list(map(int, input().split()))

maxSum = 0
for a in range(0, n):
    for b in range(a+1, n):
        for c in range(b+1, n):
            result = card[a]+card[b]+card[c]
            if result <= m:
                maxSum = max(maxSum, result)
print(maxSum)
```

```python
n = int(input())
stack = []
currentPush = 0
data = []
for i in range(n): # ê·¸ëƒ¥ ì•„ë˜ í¬ë¬¸ê³¼ í•©ì¹  ìˆ˜ ìˆìŒ
    data.append(int(input()))
result = []

for a in data:
    if a in stack:
        data = 0
        while a != data:
            data = stack.pop()
            result.append('-')
    else:
        if a >= currentPush:
            while a != currentPush: # ìœ„ ifë¬¸ê³¼ í•©ì³ì¤„ ìˆ˜ ìˆìŒ while a>currentPush:
                currentPush += 1
                stack.append(currentPush)
                result.append('+')
            stack.pop()
            result.append('-')
        else:
            result = ['NO']
            break

for i in result:
    print(i)
```

```python
n = int(input())
count = 1
stack = []
result = []

for i in range(n):
    data = int(input())
    while count <= data:
        stack.append(count)
        count += 1
        result.append('+')
    if stack[-1] == data:
        stack.pop()
        result.append('-')
    else:
        print('NO')
        exit(0) #ì•„ì˜ˆ ì¢…ë£Œ
for i in result:
    print(i)
```

# ğŸ”º 5397 í‚¤ë¡œê±°
- ë‚´ê°€ì§  ì½”ë“œì²˜ëŸ¼ ì»¤ì„œë¥¼ ë³€ë™í•˜ê³  ì¸ë±ì‹±í•´ì„œ ìˆ˜ì •í•˜ë©´ ì‹œê°„ì´ˆê³¼
- ì¸ë±ì‹±ì—ì„œ ì‹œê°„ì´ ì´ˆê³¼ë˜ëŠ” ê²ƒ ê°™ìŒ
- ì™¼ìª½ê³¼ ì˜¤ë¥¸ìª½ì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë”°ë¡œ ë‘ê³  ì»¤ì„œê°€ ì˜®ê²¨ê°ˆ ë•Œ, ë§¨ ë’¤ë¥¼ popí•´ì„œ ë‹¤ë¥¸ ê³³ìœ¼ë¡œ ì´ë™ì‹œí‚¤ê²Œ ì§œì
- ì‹œê°„ ì´ˆê³¼ ë¬¸ì œì˜ ì£¼ëœ ì›ì¸ì€ `insert(0, data)`ì™€ `pop(0)`ì´ì—ˆìŒ!

```python
testCast = int(input())
for _ in range(testCast):
    data = input()
    cursor = 0
    cnt = 0
    result = []
    for item in data:
        if item == '<':
            if cursor > 0:
                cursor -= 1
            else:
                continue
        elif item == '>':
            if cursor < cnt:
                cursor += 1
            else:
                continue
        elif item == '-':
            if cnt > 0 and cursor > 0:
                result.pop(cursor-1)
                cnt -= 1
                cursor -= 1
            else:
                continue
        else:
            result.insert(cursor,item)
            cursor += 1
            cnt += 1

    print(''.join(result))
```

```python
# ë§¨ ì•ì—ê±°ë¥¼ ì°¾ê³ , ë§¨ ì•ì— insertí•´ì£¼ëŠ” ê²Œ ì‹œê°„ì´ˆê³¼
# insert >>>>>>> append
# pop(0) >>>>>>> pop()

testCase = int(input())

for _ in range(testCase):
    data = input()
    left, right = [], []
    for item in data:
        if item == '<':
            if len(left) > 0:
                temp = left.pop()
                right.insert(0, temp) # ì—¬ê¸°ì„œ ì‹œê°„ì´ˆê³¼!!!
            else:
                continue
        elif item == '>':
            if len(right) > 0:
                temp 2= right.pop(0)
                left.append(temp) # ì—¬ê¸°ì„œ ì‹œê°„ì´ˆê³¼!!!
            else:
                continue
        elif item == '-':
            if len(left) > 0:
                left.pop()
            else:
                continue
        else:
            left.append(item)

    print(''.join(left)+''.join(right))
```

```python
testCase = int(input())

for _ in range(testCase):
    data = input()
    left, right = [], []
    for item in data:
        if item == '<':
            if len(left) > 0:
                temp = left.pop()
                right.append(temp)
            else:
                continue
        elif item == '>':
            if len(right) > 0:
                temp = right.pop()
                left.append(temp)
            else:
                continue
        elif item == '-':
            if len(left) > 0:
                left.pop()
            else:
                continue
        else:
            left.append(item)

    print(''.join(left)+''.join(reversed(right)))
```

```python
from collections import defaultdict
N = int(input())
for _ in range(N):
    F = int(input())
    graph = defaultdict(list)
    for i in range(F):
        data = input().split(' ')
        graph[data[0]].append(data[1])
        graph[data[1]].append(data[0])
        queue = []
        result = set()
        queue.append(data[0])
        while len(queue) > 0:
            temp = queue.pop()
            if temp not in result:
                queue.extend(graph[temp])
                result.add(temp)
        print(len(result))
```

```python
def union(x,y):
    rootx = find(x)
    rooty = find(y)
    if rootx != rooty: #ì´ë¯¸ ê°™ì€ ì§‘í•©ë‚´ì— ìˆëŠ” ê²½ìš°ë¥¼ ì œì™¸í•˜ê³  
        parent[rootx] = rooty
        num[rooty] += num[rootx]

def find(x):
    if parent[x] != x:
        parent[x] = find(parent[x])
    return parent[x]
        
n = int(input())

for _ in range(n):
    parent = dict()
    num = dict()
    m = int(input())
    for _ in range(m):
        x, y = input().split(' ')
        if x not in parent:
            parent[x] = x
            num[x] = 1
        if y not in parent:
            parent[y] = y
            num[y] = 1
        union(x, y)
        print(num[find(y)])
```
