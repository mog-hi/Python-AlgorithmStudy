```python
# ì•„ëž˜ëŠ” ì‹œê°„ ì´ˆê³¼
n = int(input())
isEmpty = [0 for _ in range(n+1)]
remain = []
for _ in range(n):
    temp = int(input())
    if isEmpty[temp] == 0:
        isEmpty[temp] = 1
    else:
        remain.append(temp)
        
remain.sort(reverse = True)
result = 0
for i in range(1, len(isEmpty)):
    if isEmpty[i] == 0:
        temp = remain.pop()
        result += abs(temp - i)
print(result)
```

```python
n = int(input())
crane = list(map(int, input().split()))
m = int(input())
box = list(map(int, input().split()))

crane.sort()
box.sort(reverse = True)

time = 0
if box[0] > crane[-1]:
    print(-1)
else:
    while box:
        temp = box.pop()
        for i in range(n):
            if temp <= crane[i]:
                if len(box) > 0 and i < n-1:
                    temp = box.pop()
                else:
                    break
            else:
                continue
        time += 1
    print(time)    
            
```
