# 1461. ë„ì„œê´€

```python
n,m = map(int,input().split())
book = list(map(int, input().split()))
left = []
right = []
for item in book:
    if item < 0:
        left.append(item)
    elif item > 0:
        right.append(item)

distance = []
left.sort()
for i in range(len(left)//m):
    distance.append(abs(left[m*i]))
if len(left) % m > 0:
    distance.append(abs(left[(len(left)//m)*m]))
    
right.sort(reverse = True)
for i in range(len(right)//m):
    distance.append(right[m*i]) 
if len(right) % m > 0:
    distance.append(right[(len(right)//m)*m])    
    
distance.sort()
result = distance.pop()
result += 2*sum(distance)
print(result)    
```

```python
import heapq
n = int(input())
queue = []
for _ in range(n):
    deadline, cupNoodle = map(int, input().split())
    heapq.heappush(queue, (deadline, cupNoodle))
    
current = 0
currentCup = 0
result = 0
while queue:
    temp = heapq.heappop(queue)
    if current != temp[0]:
        result += temp[1]
        current = temp[0]
        currentCup = temp[1]
    else:
        if currentCup < temp[1]:
            result -= currentCup
            result += temp[1]
            current = temp[0]
            currentCup = temp[1]
            
print(result)
```

```python
import heapq
n = int(input())
array = []
for _ in range(n):
    deadline, cupNoodle = map(int, input().split())
    array.append((deadline, cupNoodle))

array.sort()

queue = []

for i in array:
    if i[0] <= len(queue):
        heapq.heappop(queue)
        heapq.heappush(queue, i[1])
    else:
        heapq.heappush(queue, i[1])
    
print(sum(queue))
```
