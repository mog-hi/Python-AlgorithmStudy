### ìµœë‹¨ ê²½ë¡œ ì•Œê³ ë¦¬ì¦˜ : ë‹¤ìµìŠ¤íŠ¸ë¼
- distanceë°°ì—´, queue, graphë¥¼ ì¤€ë¹„í•œë‹¤.
- queueì—ì„œ êº¼ë‚¸ nodeë¥¼ cnodeë¼ê³  í• ë•Œ cnodeì™€ ì¶œë°œì  ì‚¬ì´ì˜ ê±°ë¦¬ + cnodeì™€ ë‹¤ë¥¸ ì¸ì ‘í•œ node(v)ì‚¬ì´ì˜ ê±°ë¦¬ê°€ distanceì— ì´ë¯¸ ì €ì¥ëœ ì¶œë°œì ê³¼ vì‚¬ì´ì˜ ê±°ë¦¬ë³´ë‹¤ ì‘ì„ ë•Œ, ê·¸ ê°’ì„ êµì²´

```python
import heapq
# [node for node in graph] : ì—ì„œ nodeëŠ” graphì˜ keyê°’ë“¤ ì´ë‹¤.
# for key, value in graph : í•˜ë©´ key, valueê°€ ì•Œë§ê²Œ ë“¤ì–´ê°„ë‹¤.
def dijkstra(graph, start):
    distance = {node: float('inf') for node in graph}
    distance[start] = 0
    queue = []
    heapq.heappush(queue, [start, distance[start]])
    while queue:
        currentNode, currentDis = heapq.heappop(queue)
        if distance[currentNode] < currentDis:
            continue
        for node, dis in graph[currentNode].items(): # ì§€ê¸ˆ popí•œ ì• ë‘ ì¸ì ‘í•œ ì• ë“¤, ì§€ê¸ˆ popí•œì•  ~ ì¶œë°œì  ê¹Œì§€ì˜ ê±°ë¦¬ + ì¸ì ‘í•œ ë…¸ë“œ ê¹Œì§€ì˜ ê±°ë¦¬ ë” ì‘ì€ì§€ ë¹„êµí•˜ê³  êµì²´
            if dis + currentDis < distance[node]:
                distance[node] = dis + currentDis
                heapq.heappush(queue, [node, distance[node]])
    return distance
```

```python
from collections import defaultdict
from collections import deque
n, m = list(map(int, input().split(' ')))
graph = defaultdict(dict)

Min = float('inf')
Max = 0
for _ in range(m):  
    a, b, c = list(map(int, input().split(' ')))
    if Min > c:
        Min = c
    if Max < c:
        Max = c
    graph[a][b] = c
    graph[b][a] = c

start, end = list(map(int, input().split(' ')))    

def BFS(graph, start, end, mid):
    queue = deque()
    visited = [False] * (n+1)
    queue.append(start)
    visited[start] = True
    while queue:
        node = queue.popleft()
        if node == end:
            break
        for cn, w in graph[node].items():
            if not visited[cn] and mid <= w:
                queue.append(cn)
                visited[cn] = True
    return visited[end]

maxWeight = 0
while Min <= Max:
    mid = (Min+Max) // 2
    result = BFS(graph, start, end, mid)
    if result:
        maxWeight = mid
        Min = mid + 1
    else:
        Max = mid - 1
        
print(maxWeight)

# visited ë°”ê¿ˆ 
```

# ğŸ”ºğŸ”º2250. íŠ¸ë¦¬ì˜ ë†’ì´ì™€ ë„ˆë¹„
- rootê°€ í•­ìƒ 1ì´ ì•„ë‹ ìˆ˜ ìˆìŒ ì£¼ì˜
- inorderì„ í†µí•´ íŠ¸ë¦¬ì˜ level(ë†’ì´)ì™€ ë„ˆë¹„ë¥¼ êµ¬í•  ìˆ˜ ìˆë‹¤!
- levelì„ êµ¬í•  ë•Œ, ëª¨ë“  ë…¸ë“œë¥¼ countí•˜ì§€ ì•Šê³  ìì‹ìœ¼ë¡œ ë‚´ë ¤ê°ˆ ë•Œë§Œ countí•´ì„œ levelì„ êµ¬í•˜ê³  ì‹¶ì€ê±°ë‹ˆê¹Œ -> inorderì—ì„œ levelì„ íŒŒë¼ë¯¸í„°ë¡œ ì „ë‹¬í•˜ì! 
- ë…¸ë“œ ì´ë¦„ì´ ë‹¤ ì •ìˆ˜ì¼ ë•Œ, ë°°ì—´ì˜ ì¸ë±ìŠ¤ë¥¼ í™œìš©í•´ì„œ ë”•ì…”ë„ˆë¦¬ì²˜ëŸ¼ ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤. 

```python
from collections import defaultdict
import copy

# ê·¸ë˜í”„ ì…ë ¥ë°›ê¸°
# root ê°€ í•­ìƒ 1ì´ ì•„ë‹˜ -> parent ê°’ì„ ë°›ì•„ì„œ parentê°€ ìê¸°ìì‹ ì¸ ì¦‰, rootì¸ ë…¸ë“œ ì°¾ëŠ” forë¬¸ ëŒë ¤ì„œ ì°¾ì•„ì•¼í•´
n = int(input())
graph = defaultdict(list)
parent = defaultdict(int)
for _ in range(n):
    a, b ,c = map(int,input().split(' '))
    if b != -1:
        parent[b] = a
    if c != -1:
        parent[c] = a
    graph[a] = [b, c]

# root ì°¾ê¸°
root = 1
for i in range(1, n+1): 
    if parent[i] == 0:
        root = i

# ì—´êµ¬í•˜ê¸° 
# ì—¬ê¸°ì„œ levelë„ ì•Œ ìˆ˜ ìˆë‹¤!! 
# inorderì—ì„œ left -> root -> right ë°©ë¬¸
# ì´ë•Œ, left, right ë°©ë¬¸ì—ì„œ ë™ì¼í•˜ê²Œ level + 1ë¡œ ì£¼ê¸° : ê·¸ëƒ¥ ë³€ìˆ˜ì— ê³„ì† ë”í•˜ë©´
# ë ˆë²¨ ë§ˆë‹¤ê°€ ì•„ë‹ˆë¼ ë…¸ë“œ ë§ˆë‹¤ ë”í•´ì§€ë¯€ë¡œ íŒŒë¼ë¯¸í„°ë¡œ levelì„ ë§Œë“ ë‹¤!! 
# ê° ë ˆë²¨ë§ˆë‹¤ ìµœì†Œ ë„ˆë¹„, ìµœëŒ€ ë„ˆë¹„ ì €ì¥
# ì¸ë±ìŠ¤ê°€ ë ˆë²¨ì´ ëœë‹¤ minWidth[level]ë¡œ ì €ì¥!! (dictionaryì²˜ëŸ¼ listë¥¼ í™œìš©)
minWidth = [n for i in range(n+1)] # ìµœì†Œ ê°’ì„ êµ¬í•˜ê¸° ìœ„í•´ ê°€ì¥ í°ê°’
maxWidth = [0 for i in range(n+1)] # ìµœëŒ€ ê°’ì„ êµ¬í•˜ê¸° ìœ„í•´ ê°€ì¥ ì‘ì€ ê°’
def inorder(root, level):
    global col
    if graph[root][0] != -1:
        inorder(graph[root][0], level+1)
    minWidth[level] = min(minWidth[level], col)
    maxWidth[level] = max(maxWidth[level], col)
    col += 1
    if graph[root][1] != -1:
        inorder(graph[root][1], level+1)
        
col = 1
inorder(root, 1)

result = maxWidth[1] - minWidth[1] + 1
resultLevel = 1
for i in range(2, n+1):
    width = maxWidth[i] - minWidth[i] + 1
    if width > result:
        result = width
        resultLevel = i
print(resultLevel, result)
```

```python
import heapq

n, m = map(int, input().split())

# ë°°ì—´ì„ ë”•ì…”ë„ˆë¦¬ ì²˜ëŸ¼ ì“°ê¸° (ì¸ë±ìŠ¤ ê°’ìœ¼ë¡œ)
# ë°©í–¥ì´ ìˆëŠ” ê·¸ë˜í”„ì˜ ê²½ìš° ë‚´ê°€ ê°ˆ ìˆ˜ ìˆëŠ” ë…¸ë“œë¥¼ ê¸°ë¡í•œë‹¤.
graph = [[] for i in range(n+1)] # ì—°ê²° ë˜ì–´ ìˆëŠ” ë…¸ë“œ 
indegree = [0 for i in range(n+1)] # ì§„ì…ì°¨ìˆ˜ 

for _ in range(m):
    x, y = map(int, input().split()) # x -> y (xë‹¤ìŒì— yê°€ ì‹¤í–‰ë¼ì•¼ í•˜ë‹ˆê¹Œ)
    graph[x].append(y) # xì— yì—°ê²° (xê°€ ë¨¼ì €)
    indegree[y] += 1 # yëŠ” x ì´í›„ì— í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ ì§„ì…ì°¨ìˆ˜ê°€ 1 

queue = []
result = []
for i in range(1, n+1):
    if indegree[i] == 0:
        heapq.heappush(queue, i)
        
while queue:
    temp = heapq.heappop(queue)
    result.append(temp)
    for item in graph[temp]:
        indegree[item] -= 1
        if indegree[item] == 0:
            heapq.heappush(queue, item)
    graph[temp] = []
    
for item in result:
    print(item, end = ' ')
```
