# 1260. DFSì™€ BFS
- ìŠ¤íƒë§ê³  ì¬ê·€í•¨ìˆ˜ë¥¼ ì´ìš©í•œ dfsë„ ê¸°ì–µí•´ë‘ê¸°!!
- ê°„ì„  ì €ì¥ ìˆœì„œì— ë”°ë¼ ë‹¬ë¼ì§€ê¸° ë•Œë¬¸ì— ê°„ì„  ìˆœì„œ ì •ë ¬í•œ í›„ íƒìƒ‰í•´ì•¼í•œë‹¤.

```python
from collections import deque, defaultdict
n, m, start = map(int, input().split())

graph = defaultdict(list)
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
# BFS
visited = [0 for _ in range(n+1)]
queue = deque()
queue.append(start)
bfs = []
while queue:
    temp = queue.popleft()
    if visited[temp] == 0:
        bfs.append(temp)
        visited[temp] = 1
        graph[temp].sort()
        queue.extend(graph[temp])
        
# DFS
# stackì„ ì´ìš©í•´ë„ ë˜ì§€ë§Œ ì¬ê·€í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ê² ìŒ
# ìê¸°ë‘ ì—°ê²°ëœì•  dfs í˜¸ì¶œ, í˜¸ì¶œ 
visited = [0 for _ in range(n+1)]
def dfs(start):
    print(start, end = ' ')
    visited[start] = 1
    for item in graph[start]:
        if visited[item] == 0:
            dfs(item)
            

dfs(start)
print()
for item in bfs:
    print(item, end = ' ')
    
```

# 1012. ìœ ê¸°ë† ë°°ì¶”
- ì—¬ê¸°ì„  bfsìˆœì„œê°€ ì¤‘ìš”í•œê²Œ ì•„ë‹ˆë‹ˆê¹Œ ê·¸ëƒ¥ stack, dfsë¡œ êµ¬í˜„í•˜ì
- dequeë¥¼ ì‚¬ìš©í•˜ëŠ”ê²Œ pop(0)ë³´ë‹¨ ì‹œê°„ì´ ëœê±¸ë¦¬ì§€ë§Œ pop()ë§Œí¼ì€ ì•„ë‹ˆë‹¤!

```python
# from collections import deque
testCase = int(input())
for _ in range(testCase):
    m, n, k = map(int, input().split())
    table = [[0] * n for _ in range(m)]
    array = []
    for _ in range(k):
        a, b = map(int, input().split())
        array.append((a,b))
        table[a][b] = 1

    def bfs(x, y):
        queue = []
        queue.append((x,y))
        while queue:
            a, b = queue.pop()
            table[a][b] = 0
            if a-1 >= 0 and table[a-1][b] == 1:
                queue.append((a-1, b))
            if a+1 < m and table[a+1][b] == 1:
                queue.append((a+1, b))
            if b-1 >= 0 and table[a][b-1] == 1:
                queue.append((a, b-1))
            if b+1 < n and table[a][b+1] == 1:
                queue.append((a, b+1))
    cnt = 0
    for x, y in array:
        if table[x][y] == 1:
            bfs(x, y)
            cnt += 1
    print(cnt)
```

# 10282 í•´í‚¹
- ê°„ì„ ì— weightê°€ ì£¼ì–´ì§„ë‹¤ 
- ê°€ì¥ ê±°ë¦¬ê°€ ë¨¼ ë…¸ë“œì˜ ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” ê²ƒ -> ìµœë‹¨ ê²½ë¡œ 
- ì´ëŸ° ê²½ìš°ì˜ íƒìƒ‰ì€ ë‹¤ìµìŠ¤íŠ¸ë¼ë¥¼ ì´ìš©í•œë‹¤. 

```python
import heapq
test_case = int(input())
for _ in range(test_case):
    n, d, c = map(int, input().split())
    graph = [[] for _ in range(n+1)]
    for i in range(d):
        a, b, s = map(int, input().split())
        graph[b].append((s, a))

    queue = []
    heapq.heappush(queue, (0, c))
    distances = [float('inf') for _ in range(n+1)]
    distances[c] = 0
    while queue:
        cS, cNode = heapq.heappop(queue)
        if cS <= distances[cNode]:
            for s, a in graph[cNode]:
                dis = cS + s
                if distances[a] > dis:
                    distances[a] = dis
                    heapq.heappush(queue, (dis, a))

    cnt = 0
    time = 0
    for item in distances:
        if item != float('inf'):
            cnt += 1
            if time < item:
                time = item
    print(cnt, time)
```
