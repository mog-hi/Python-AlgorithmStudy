# ğŸ”ºğŸ”º12865. í‰ë²”í•œ ë°°ë‚­
- ë‹¤ì´ë‚˜ë¯¹ í”„ë¡œê·¸ë˜ë°ìœ¼ë¡œ êµ¬í˜„í•œë‹¤.
- ë‹¤ì´ë‚˜ë¯¹ì€ ëŒ€ë¶€ë¶„  ë°°ì—´, ì´ì°¨ì› ë°°ì—´ì„ ì´ìš©í•œë‹¤.
- ìë£Œ êµ¬ì¡° ì„¤ê³„ê°€ ê°€ì¥ ì–´ë µë‹¤!
- ì´ ë¬¸ì œìœ í˜•ì€ ì™¸ì›Œë‘˜ê²ƒ!
- ëª¨ë“  ì§ì— ëŒ€í•´ì„œ ìµœëŒ€ì¹˜ êµ¬í•˜ê¸° (i, ê°í–‰)
- ë¬´ê²Œì˜ í•©ì€ j
- `dp[i][j]`ëŠ” ë¬´ê°œê°€ jì¼ ë•Œ ìµœëŒ€ ê°€ì¹˜ 
- i <= n
- j <= k
<img src="./12865.PNG" />
<img src="./12865_1.PNG" />

```python
n, k = map(int, input().split())
# dpì„ ì–¸ 
dp = [[0] * (k+1) for _ in range(n+1)]
for i in range(1, n+1):
    weight, value = map(int, input().split())
    for j in range(1, k+1):
        if j <weight:
            dp[i][j] = dp[i-1][j]
        else:
            dp[i][j] = max(dp[i-1][j], dp[i-1][j-weight] + value)
print(dp[n][k])
```

# ğŸ”ºğŸ”º9251. LCS

- êµ¬ì¡°ëŠ” ì˜ ì¡ì•˜ìœ¼ë‚˜ ì•„ì´ë””ì–´ê°€ í‹€ë ¸ë‹¤
- ëŒ€ê°ì„  ìœ„ì— ìˆëŠ” ì• ì™€ ë¹„êµí•œë‹¤ëŠ” ê²ƒì„ ì•Œì•„ëƒˆìœ¼ë©´, ë°”ë¡œ ëŒ€ê°ì„  ìœ„ê°€ ì•„ë‹ ê²½ìš°ë¥¼ ìœ„í•´ì„œ, ë’¤ì— ê²ƒë“¤ì„ ì±„ì›Œ ë„£ì–´ì•¼ í•¨ë„ ì•Œì•„ì•¼í•œë‹¤.
- pointëŠ” `dp[i][j] = max(dp[i][j-1], dp[i-1][j])`

```python
# ë‚´ê°€ì§  í‹€ë¦° ì½”ë“œ
array = input()
n = len(array)
array2 = input()
k = len(array2)

dp = [[0] * n for _ in range(k+1)]
for i in range(1, k+1):
    Max = 0
    for j in range(n):
        if array2[i-1] == array[j]:
            dp[i][j] = Max + 1
        else:
            dp[i][j] = dp[i-1][j]
            Max = max(Max, dp[i-1][j])

result = 0
for item in dp[k]:
    result = max(result, item)
print(result)
```

```python
x = input()
y = input()

dp = [[0]*(len(y)+1) for _ in range(len(x)+1)]

for i in range(1, len(x)+1):
    for j in range(1, len(y)+1):
        if x[i-1] == y[j-1]:
            dp[i][j] = dp[i-1][j-1] + 1
        else:
            dp[i][j] = max(dp[i][j-1], dp[i-1][j])
print(dp[len(x)][len(y)])
```

```python
n, s, m = map(int, input().split())
v = list(map(int, input().split()))
dp = [[] for _ in range(n+1)]
dp[0].append(s)
for i in range(1, n+1):
    for j in range(len(dp[i-1])):
        vol1 = dp[i-1][j] - v[i-1]
        if vol1 > 0 and vol1 < m:
            dp[i].append(vol1)
        vol1 = dp[i-1][j] - v[i-1]
        if vol1 > 0 and vol1 < m:
            dp[i].append(vol1)
```
